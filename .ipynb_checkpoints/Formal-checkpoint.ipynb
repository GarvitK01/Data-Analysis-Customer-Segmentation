```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')
from collections import Counter
```

## Data Exploration and Preparation

```python
customers = pd.read_csv("./Customers.csv")
```

```python
customers.isna()
#Dropping any records with NaN value
customers = customers.dropna()
```

```python
customers.head(10)
```

```python
customers = customers.drop(labels = "ID", axis = 1)
```

```python
customers
```

```python
# Converting Cateogrical into one hot encoding
from sklearn.preprocessing import OneHotEncoder

binary = ["Ever_Married", "Graduated", "Gender"]
category = ["Profession", "Spending_Score"]

def convert_binary(fields, df):
    
    for field in fields:
        
        df[field] = df[field].apply(lambda X: 1 if X == "Yes" or X == "Male" else 0)
        
    return df

def transform_category(fields, df):
    
    encoder = OneHotEncoder(handle_unknown='ignore')
    encoded = []
    for field in fields:
        
        encoded.append(pd.DataFrame(encoder.fit_transform(df[[field]]).toarray(), columns =list(Counter(df[field].to_list()).keys())))
        
    encoded_df = pd.concat([encoded[0], encoded[1]], axis = 1)   
    
    return encoded_df
    
```

```python
customer_1 = convert_binary(binary, customers)
```

```python
encoded_fields = transform_category(category, customer_1)
```

```python
# Dropping features
customer_1.drop(labels = ['Profession', 'Spending_Score'], axis = 1, inplace = True)
```

```python
customer_1.drop(labels = ['Var_1'], axis = 1, inplace = True)
```

```python
customer_1 = customer_1.reset_index()
```

```python
customer_1.drop(labels = ['index'], axis = 1, inplace = True)
```

```python
customer_1 = pd.concat([customer_1, encoded_fields], axis = 1)
```

```python
customer_1
```

## Scaling Data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
categories = ["Age", "Work_Experience", "Family_Size"]

def scaling(fields, df):
    
    for field in fields:
        df[field] = scaler.fit_transform(df[[field]])
        
    return df

scaled_df = scaling(categories, customer_1)
```

```python
scaled_df
```

## T-SNE Embeddings 2D

```python
from sklearn.manifold import TSNE
embeding_2d = TSNE(n_components=2, perplexity = 2).fit_transform(scaled_df)
```

```python
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE

kmean = KMeans(n_clusters = 6).fit(embeding_2d)
labels = kmean.predict(embeding_2d)

def plot_2d():
    color_list = ['blue', 'yellow', 'orange', 'green', 'red', 'violet', 'pink']
    fig = plt.figure()
    plt.show()
    for i in range(7):
        
        label1 = embeding_2d[labels == i]
        plt.scatter(label1[:,0] , label1[:,1] , color = color_list[i])
        
    plt.show();
    
plot_2d()
```

## Principal Component Analysis

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE

pca = PCA(n_components=2)
embedding_2d = pca.fit_transform(scaled_df)
kmean = KMeans(n_clusters = 5).fit(embedding_2d)
labels = kmean.predict(embedding_2d)

def plot_2d():
    color_list = ['blue', 'yellow', 'orange', 'green', 'red', 'violet', 'pink']
    fig = plt.figure()
    plt.show()
    for i in range(6):
        
        label1 = embedding_2d[labels == i]
        plt.scatter(label1[:,0] , label1[:,1] , color = color_list[i])
        
    plt.show();
    
plot_2d()





embeding_3d = PCA(n_components=3, svd_solver = 'full').fit_transform(scaled_df)
clustering = KMeans(6).fit(embeding_3d)
labels = clustering.predict(embeding_3d)

fig = plt.figure(figsize = (8, 8))
ax = plt.axes(projection="3d")

def plot_bs_3d(X_embedded, labels):
    color_list = ['blue', 'yellow', 'orange', 'green', 'red', 'violet', 'pink']
    for i in range(6):
        
        label_i = X_embedded[labels == i]
        x = label_i[:, 0]
        y = label_i[:, 1]
        z = label_i[:, 2]
        
        ax.scatter3D(x, y, z, c= color_list[i])
        
    plt.show()
plot_bs_3d(embeding_3d, labels)
```

```python
customer_1["Segment"] = labels
```

```python
customer_1.to_csv("Scaled_PCA.csv")
```
